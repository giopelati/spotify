## Etapa 0: **Data Loading**
- Consolidar múltiplos arquivos de streaming em um único dataset.
- Garantir a correta identificação de cada usuário.

```python
import pandas as pd
from pathlib import Path
```

```python
base_path = Path("../dados")

all_dfs = []

for user_folder in base_path.iterdir():
    if user_folder.is_dir():
        user_name = user_folder.name
        
        for file in user_folder.glob("Streaming_History_Audio_*.json"):
            df = pd.read_json(file)
            df["username"] = user_name
            all_dfs.append(df)

df = pd.concat(all_dfs, ignore_index=True)
```

## Etapa 2: **Início de Data Preparation**
- Remover registros que não correspondem a músicas (ex.: podcasts).
- Criar o track_id.
- Limpar e renomear as colunas.
- Criar dataset limitado para balancear o número de músicas por usuário.

```python
df = df[df["spotify_track_uri"].notna()]
```

```python
df["track_id"] = df["spotify_track_uri"].str.split(":").str[-1]
```

```python
df = df[['ts', 'ms_played', 'master_metadata_track_name', 'master_metadata_album_artist_name', 'master_metadata_album_album_name', 'username', 'track_id']]
```

```python
df = df.rename(columns={
    'master_metadata_track_name': 'track_name',
    'master_metadata_album_artist_name': 'artist_name',
    'master_metadata_album_album_name': 'album_name'
})
```

```python
df.head()
```

```python
N = 10000

df_limited = (
    df
    .sort_values("ts")
    .groupby("username", group_keys=False)
    .tail(N)
)
```
