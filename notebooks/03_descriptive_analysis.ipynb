```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import plotly.express as px
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Etapa 0: Carregamento do dataset processado
# Carrega o CSV já tratado/retornado por etapas anteriores
df = pd.read_csv('../datasets/return_corrected.csv')
df.head()
```

```python
# Etapa 1: Filtragem inicial e definição de variáveis de interesse
# Cria variável binária 'heard' (1 = música ouvida > 30s, 0 = não)
df["heard"] = (df["ms_played"] > 30000).astype(int)

# Mantém apenas as reproduções que representam músicas efetivamente ouvidas
df_heard = df[df["heard"] == 1].copy()

# Etapa 2: Definição das features musicais utilizadas no perfil
feature_cols = [
    "acousticness",
    "danceability",
    "energy",
    "instrumentalness",
    "liveness",
    "loudness",
    "speechiness",
    "valence"
]

# Etapa 3: Perfil médio ponderado por tempo ouvido (ms_played)
# Ponderamos cada feature pelo tempo que o usuário ouviu a música
def weighted_profile(x):
    return np.average(
        x[feature_cols],
        axis=0,
        weights=x["ms_played"]
    )

# Aplica o cálculo do perfil ponderado para cada usuário
user_profiles = (
    df_heard
    .groupby("user_name")
    .apply(weighted_profile)
)

# Converte o resultado em DataFrame com nomes de colunas explícitos
user_profiles = pd.DataFrame(
    user_profiles.tolist(),
    index=user_profiles.index,
    columns=feature_cols
)

# Etapa 4: Escala utilizada para cálculo de diversidade (Min-Max) e agregação
scaler = MinMaxScaler()
df_heard_scaled = df_heard.copy()
df_heard_scaled[feature_cols] = scaler.fit_transform(df_heard[feature_cols])

# Diversidade musical = média das stds das features por usuário (mesma escala)
user_diversidade = (
    df_heard_scaled
    .groupby("user_name")[feature_cols]
    .std()
    .mean(axis=1)
)
user_profiles["diversidade_musical"] = user_diversidade

# Etapa 5: Interpretação do perfil via percentis (Top/Bottom quartis)
def interpretar_perfil_avancado(row, df_ref):
    perfil = []

    # Energia: top/bottom quartis
    if row["energy"] > df_ref["energy"].quantile(0.75):
        perfil.append("Extremamente Energético")
    elif row["energy"] < df_ref["energy"].quantile(0.25):
        perfil.append("Gosta de Músicas Calmas")

    # Dançabilidade
    if row["danceability"] > df_ref["danceability"].quantile(0.75):
        perfil.append("Muito Dançante")

    # Valence (mood)
    if row["valence"] > df_ref["valence"].quantile(0.75):
        perfil.append("Vibes Positivas/Alegres")
    elif row["valence"] < df_ref["valence"].quantile(0.25):
        perfil.append("Vibes Melancólicas/Densas")

    # Instrumentalness (corte mais restritivo)
    if row["instrumentalness"] > df_ref["instrumentalness"].quantile(0.85):
        perfil.append("Fã de Instrumentais")

    # Diversidade musical
    if row["diversidade_musical"] > df_ref["diversidade_musical"].quantile(0.75):
        perfil.append("Gosto Altamente Eclético")
    elif row["diversidade_musical"] < df_ref["diversidade_musical"].quantile(0.25):
        perfil.append("Gosto Muito Específico/Focado")

    return ", ".join(perfil) if perfil else "Gosto Muito Próximo à Média Geral"

# Aplica a interpretação ao DataFrame de perfis
user_profiles["tipo_preferido"] = user_profiles.apply(
    lambda row: interpretar_perfil_avancado(row, user_profiles),
    axis=1
)
```

```python
# Etapa: Visualização rápida dos perfis gerados
pd.set_option('display.max_colwidth', None)
display(user_profiles.head())
```

```python
cols_comparacao = [
    "acousticness", "danceability", "energy", 
    "instrumentalness", "liveness", "speechiness", "valence"
]

fig = px.imshow(
    user_profiles[cols_comparacao],
    labels=dict(x="Características da Música", y="Usuários", color="Intensidade"),
    x=cols_comparacao,
    y=user_profiles.index,
    color_continuous_scale="Greens",
    text_auto=".2f"
)

fig.update_layout(
    title=dict(text="Comparação Direta: Quem é quem nas Features Musicais?", font=dict(size=18)),
    xaxis_tickangle=-45,
    margin=dict(t=60, b=60, l=60, r=60)
)

fig.show()
```

```python
# Etapa: Visualização - Scatter plot Energia x Valence
fig = px.scatter(
    user_profiles,
    x="energy",
    y="valence",
    color="tipo_preferido",
    size="diversidade_musical",
    hover_name=user_profiles.index,
    title="Mapa de Humor dos Usuários (Energia vs. Valence)"
)

fig.update_layout(shapes=[
    dict(type='line', x0=0.5, y0=0, x1=0.5, y1=1, line=dict(dash='dash', color='gray')),
    dict(type='line', x0=0, y0=0.5, x1=1, y1=0.5, line=dict(dash='dash', color='gray'))
])

fig.show()
```

```python
# Etapa: Similaridade entre perfis
# 1) Padroniza (Z-Score) para garantir equalização das variâncias
scaler = StandardScaler()
user_profiles_zscore = pd.DataFrame(
    scaler.fit_transform(user_profiles[feature_cols]),
    index=user_profiles.index,
    columns=feature_cols
)

# 2) Matriz de similaridade por cosseno
matrix_z = cosine_similarity(user_profiles_zscore)
df_similidade_z = pd.DataFrame(matrix_z, index=user_profiles.index, columns=user_profiles.index)

# 3) Heatmap interativo da similaridade
fig = px.imshow(
    df_similidade_z,
    text_auto=".1%",
    color_continuous_scale="RdYlGn",
    color_continuous_midpoint=0,
    labels=dict(color="Similaridade"),
    aspect="auto"
)

fig.update_layout(
    title=dict(text="Distância Real: O quanto fugimos do padrão do grupo?", font=dict(size=18)),
    xaxis_title="Usuários",
    yaxis_title="Usuários",
    margin=dict(t=60, b=40, l=60, r=60)
)

fig.show()
```
