## Importação das Bibliotecas

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

## Carregamento do Dataset e Criação da Variável Target

- Carregamos o dataset `return_corrected.csv`
- Criamos a variável alvo `heard`:
  - 1 se a música foi escutada por mais de 30 segundos (`ms_played > 30000`)
  - 0 caso contrário
- Definimos as features musicais de interesse
- Separamos uma base mínima para posterior split treino/teste

```python
# Etapa 0: Carregamento do dataset processado
# Carrega o CSV já tratado/retornado por etapas anteriores
df = pd.read_csv('../datasets/return_corrected.csv')
df.head()
```

```python
# Etapa 1: Filtragem inicial e definição de variáveis de interesse
# Cria variável binária 'heard' (1 = música ouvida > 30s, 0 = não)
df["heard"] = (df["ms_played"] > 30000).astype(int)

# Mantém apenas as reproduções que representam músicas efetivamente ouvidas
df_heard = df[df["heard"] == 1].copy()

# Converte a coluna 'ts' para o formato de data real
df_heard["ts"] = pd.to_datetime(df_heard["ts"])

# Extrai o dia da semana e o nome para ordenação e visualização
# (0=Segunda, 6=Domingo)
df_heard["dia_semana_num"] = df_heard["ts"].dt.dayofweek
df_heard["dia_nome"] = df_heard["ts"].dt.day_name()

# Etapa 2: Definição das features musicais utilizadas no perfil
feature_cols = [
    "acousticness",
    "danceability",
    "energy",
    "instrumentalness",
    "liveness",
    "loudness",
    "speechiness",
    "valence"
]

# Etapa 3: Perfil médio ponderado por tempo ouvido (ms_played)
# Ponderamos cada feature pelo tempo que o usuário ouviu a música
def weighted_profile(x):
    return np.average(
        x[feature_cols],
        axis=0,
        weights=x["ms_played"]
    )

# Aplica o cálculo do perfil ponderado para cada usuário
user_profiles = (
    df_heard
    .groupby("user_name")
    .apply(weighted_profile)
)

# Converte o resultado em DataFrame com nomes de colunas explícitos
user_profiles = pd.DataFrame(
    user_profiles.tolist(),
    index=user_profiles.index,
    columns=feature_cols
)

# Etapa 4: Escala utilizada para cálculo de diversidade (Min-Max) e agregação
scaler = MinMaxScaler()
df_heard_scaled = df_heard.copy()
df_heard_scaled[feature_cols] = scaler.fit_transform(df_heard[feature_cols])

# Diversidade musical = média do desvio padrão (std) das features por usuário
user_diversidade = (
    df_heard_scaled
    .groupby("user_name")[feature_cols]
    .std()
    .mean(axis=1)
)
user_profiles["diversidade_musical"] = user_diversidade

# Etapa 5: Interpretação do perfil via percentis (Top/Bottom quartis)
def interpretar_perfil_avancado(row, df_ref):
    perfil = []

    # Energia: top/bottom quartis
    if row["energy"] > df_ref["energy"].quantile(0.75):
        perfil.append("Extremamente Energético")
    elif row["energy"] < df_ref["energy"].quantile(0.25):
        perfil.append("Gosta de Músicas Calmas")

    # Dançabilidade
    if row["danceability"] > df_ref["danceability"].quantile(0.75):
        perfil.append("Muito Dançante")

    # Valence (humor/vibe)
    if row["valence"] > df_ref["valence"].quantile(0.75):
        perfil.append("Vibes Positivas/Alegres")
    elif row["valence"] < df_ref["valence"].quantile(0.25):
        perfil.append("Vibes Melancólicas/Densas")

    # Instrumentalness (corte mais restritivo)
    if row["instrumentalness"] > df_ref["instrumentalness"].quantile(0.85):
        perfil.append("Fã de Instrumentais")

    # Diversidade musical
    if row["diversidade_musical"] > df_ref["diversidade_musical"].quantile(0.75):
        perfil.append("Gosto Altamente Eclético")
    elif row["diversidade_musical"] < df_ref["diversidade_musical"].quantile(0.25):
        perfil.append("Gosto Muito Específico/Focado")

    return ", ".join(perfil) if perfil else "Gosto Muito Próximo à Média Geral"

# Aplica a interpretação ao DataFrame de perfis
user_profiles["tipo_preferido"] = user_profiles.apply(
    lambda row: interpretar_perfil_avancado(row, user_profiles),
    axis=1
)

# Etapa 6: Clusterização de usuários via K-Means (StandardScaler)
# Normalizamos os dados para que colunas com escalas maiores (ex: loudness) não enviesem o modelo
scaler_kmeans = StandardScaler()
user_profiles_scaled = scaler_kmeans.fit_transform(user_profiles[feature_cols])

# Define a quantidade de grupos e inicializa o modelo K-Means
n_clusters = 3 
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)

# Atribui cada usuário a um cluster para análise de segmentação
user_profiles['cluster'] = kmeans.fit_predict(user_profiles_scaled)

print("Clusters criados com sucesso!")
```

## Visualização dos Perfis por Cluster (Lollipop Chart)

```python
# Etapa 7: Preparação dos dados para visualização (Agregação por Cluster)
# Calculamos a média de cada feature por cluster e reorganizamos o formato (stack)
df_lollipop = user_profiles.groupby('cluster')[feature_cols].mean().stack().reset_index()
df_lollipop.columns = ['cluster', 'feature', 'value']

# Configuração da figura para múltiplos subplots
plt.figure(figsize=(14, 8))

for i in range(n_clusters):
    plt.subplot(2, 2, i+1)
    
    # Filtra os dados específicos do cluster atual
    data = df_lollipop[df_lollipop['cluster'] == i]
    
    # Desenha as linhas horizontais (caule do pirulito)
    plt.hlines(y=data['feature'], xmin=0, xmax=data['value'], color='lightgrey', alpha=0.7)
    
    # Desenha os pontos (topo do pirulito) com cores distintas por cluster
    plt.scatter(data['value'], data['feature'], color=['#1DB954', '#4682B4', '#E91E63', '#FF9800'][i], s=100)
    
    plt.title(f'Perfil Médio: Cluster {i}', fontweight='bold')
    plt.xlim(0, 1)  # Mantém a escala fixa (0 a 1) para facilitar a comparação visual
    sns.despine()

# Ajusta o layout para evitar sobreposição de elementos
plt.tight_layout()
plt.show()
```

## Comparação de Atributos Médios e Diversidade por Cluster (Heatmap)

```python
# Etapa 8: Resumo estatístico das features e diversidade por cluster
# Agrupa os perfis pelo cluster e calcula a média de todos os atributos musicais
cluster_summary = user_profiles.groupby('cluster')[feature_cols + ['diversidade_musical']].mean()

# Configuração da visualização via Heatmap (Mapa de Calor)
plt.figure(figsize=(12, 5))

# Gera o mapa de calor para identificar padrões de intensidade entre os grupos
sns.heatmap(
    cluster_summary, 
    annot=True,          # Exibe os valores numéricos em cada célula
    cmap='YlGnBu',       # Paleta de cores (Amarelo, Verde e Azul)
    fmt=".2f",           # Formatação com duas casas decimais
    linewidths=.5        # Espaçamento sutil entre os blocos
)

plt.title('Intensidade Média de cada Atributo por Cluster', fontsize=15, fontweight='bold')
plt.show()
```

## Distribuição de Preferências Musicais por Cluster (Treemap)

```python
# Etapa 9: Agrupamento de dados para o Treemap
df_tree = user_profiles.groupby(['cluster', 'tipo_preferido']).size().reset_index(name='contagem')

fig = px.treemap(
    df_tree, 
    path=['cluster', 'tipo_preferido'], 
    values='contagem',
    color='cluster',
    color_continuous_scale='Portland',
    title='Relação entre Grupos Matemáticos e Perfis Interpretados'
)

# Ajustes de visibilidade do texto
fig.update_traces(
    textinfo="label+value", # Mostra o nome e a quantidade
    textfont_size=16,       # Aumenta o tamanho da fonte
    hovertemplate='<b>%{label}</b><br>Quantidade: %{value}'
)

# Garante que o texto tenha um tamanho mínimo legível
fig.update_layout(
    title_font_size=22,
    margin=dict(t=60, l=15, r=15, b=15),
    uniformtext=dict(minsize=12, mode='hide') # Esconde o texto se for menor que 12, evitando letras minúsculas
)

fig.show()
```
