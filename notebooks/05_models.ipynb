# Modelo Preditivo - K-Nearest Neighbors (KNN)


## Importação das Bibliotecas

Nesta etapa importamos todas as bibliotecas necessárias para:

- Manipulação de dados (Pandas e NumPy)
- Construção do modelo KNN
- Padronização das variáveis
- Criação de pipeline
- Validação cruzada com GridSearch
- Avaliação do modelo com múltiplas métricas


```python
from sklearn.model_selection import train_test_split, GridSearchCV  
from sklearn.preprocessing import StandardScaler                      
from sklearn.neighbors import KNeighborsClassifier                   
from sklearn.pipeline import Pipeline                                
from sklearn.metrics import roc_auc_score                             
import pandas as pd
import numpy as np
from sklearn.metrics import (
    accuracy_score,           
    precision_score,          
    recall_score,             
    f1_score,                 
    confusion_matrix,         
    classification_report,    
    average_precision_score  
)
```

## Carregamento do Dataset e Criação da Variável Target

- Carregamos o dataset `dados_musicas.csv`
- Criamos a variável alvo `heard`:
  - 1 se a música foi escutada por mais de 30 segundos (`ms_played > 30000`)
  - 0 caso contrário
- Definimos as features musicais de interesse
- Separamos uma base mínima para posterior split treino/teste

```python

# Configurações gerais
RANDOM_STATE = 42

# Features musicais
feature_cols = [
    "acousticness",
    "danceability",
    "energy",
    "instrumentalness",
    "liveness",
    "loudness",
    "speechiness",
    "valence"
]
# Carregar dataset
df = pd.read_csv("dados_musicas.csv")

# Checagem mínima de colunas
required_cols = ["ms_played", "user_name", "track_id"] + feature_cols
missing = [c for c in required_cols if c not in df.columns]
if missing:
    raise ValueError(f"Colunas ausentes no dataset: {missing}")

# Criação da variável alvo
df["heard"] = (df["ms_played"] > 30000).astype(int)

# Base mínima para o split
base_cols = ["user_name", "track_id", "ms_played"] + feature_cols + ["heard"]
df_base = df[base_cols].copy()

df_base.head()
```

## Divisão Treino e Teste 
- Primeiro dividimos a base em treino (80%) e teste (20%), estratificando por `heard`
- Depois calculamos o perfil médio de cada usuário

```python
# Split treino e teste com estratificação
df_train, df_test = train_test_split(
    df_base,
    test_size=0.2,
    stratify=df_base["heard"],
    random_state=RANDOM_STATE
)

df_train.shape, df_test.shape
```

## Criação do Perfil Médio dos Usuários

- Consideramos somente músicas efetivamente ouvidas (`heard = 1`) no **treino**
- Calculamos a média das features musicais por `user_name`
- Renomeamos as colunas com prefixo `user_` para indicar que fazem parte do perfil do usuário

```python
# Filtrar apenas músicas ouvidas no treino
df_heard_train = df_train[df_train["heard"] == 1]

# Perfil médio de cada usuário (apenas treino)
user_profiles = (
    df_heard_train
    .groupby("user_name")[feature_cols]
    .mean()
)

# Renomear colunas para identificar que são do usuário
user_profiles.columns = [f"user_{c}" for c in feature_cols]

user_profiles.head()
```

## Engenharia de Features: Função de Construção do Dataset de Modelo

Para cada subconjunto (treino e teste):

- Fazemos `merge` do perfil de usuário (`user_profiles`) com o dataframe
- Preenchemos perfis ausentes (usuários que só aparecem no teste) com a média global da feature
- Criamos features de diferença absoluta entre a música e o perfil do usuário:
  - `diff_acousticness`, `diff_danceability`, etc.
- Retornamos:
  - `X`: matriz de features do modelo
  - `y`: variável alvo (`heard`)
  - `model_features`: lista da ordem final das features

```python
def build_model_df(df_subset, user_profiles, feature_cols):
    """
    Constrói o dataframe de modelagem com:
    - features da música
    - perfil médio do usuário
    - diferenças absolutas entre música e perfil
    """
    # Merge com perfis de usuário
    df_model = df_subset.merge(
        user_profiles,
        left_on="user_name",
        right_index=True,
        how="left"
    )

    # Preencher perfis ausentes (usuários sem histórico de "heard=1" no treino)
    for col in feature_cols:
        user_col = f"user_{col}"
        # Preenche com a média global da própria coluna de música
        df_model[user_col] = df_model[user_col].fillna(df_model[col].mean())

    # Criar features de diferença absoluta
    for col in feature_cols:
        df_model[f"diff_{col}"] = abs(
            df_model[col] - df_model[f"user_{col}"]
        )

    # Lista final de features
    model_features = (
        feature_cols +
        [f"user_{c}" for c in feature_cols] +
        [f"diff_{c}" for c in feature_cols]
    )

    X = df_model[model_features]
    y = df_model["heard"]

    return X, y, model_features
```

```python
# Construir dataset de modelagem para treino e teste
X_train, y_train, model_features = build_model_df(
    df_train, user_profiles, feature_cols
)
X_test, y_test, _ = build_model_df(
    df_test, user_profiles, feature_cols
)

X_train.shape, X_test.shape
```

## Pipeline

Criamos um `Pipeline` com:

1. `StandardScaler` para padronização das variáveis
2. `KNeighborsClassifier` como modelo base

Usamos `GridSearchCV` para testar combinações de:

- Número de vizinhos (`n_neighbors`)
- Peso dos vizinhos (`weights`)
- Métrica de distância (`metric`)

Critério de avaliação na validação cruzada: ROC AUC.

```python
# Definição do pipeline
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("knn", KNeighborsClassifier())
])

# Grade de parâmetros para busca
param_grid = {
    "knn__n_neighbors": [10, 20, 40, 60],
    "knn__weights": ["uniform", "distance"],
    "knn__metric": ["euclidean", "manhattan"]
}

# GridSearch com validação cruzada
grid = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring="roc_auc",
    n_jobs=-1
)

grid.fit(X_train, y_train)

best_model = grid.best_estimator_

print("Melhores parâmetros:", grid.best_params_)
print("Melhor ROC AUC (validação cruzada):", grid.best_score_)
```

## Funções para Gerar Ranking de Usuários para uma Música

Para um dado `track_id`:

- Obtemos as características médias da música
- Para cada usuário:
  - Montamos o vetor de entrada contendo:
    - features da música (`acousticness`, `danceability`, etc.)
    - features do perfil do usuário (`user_acousticness`, ...)
    - diferenças absolutas (`diff_*`)
  - Calculamos a probabilidade de `heard = 1` via `best_model.predict_proba`
- Ordenamos os usuários pela probabilidade (decrescente) e retornamos um ranking

```python
def build_row_for_track_user(track_features, user_vector, feature_cols, model_features):
    """
    Monta uma linha de features (DataFrame 1xN) para um par (música, usuário).
    """
    row_dict = {}
    for col in feature_cols:
        # Feature da música
        row_dict[col] = track_features[col]
        # Perfil médio do usuário
        row_dict[f"user_{col}"] = user_vector[f"user_{col}"]
        # Diferença absoluta
        row_dict[f"diff_{col}"] = abs(track_features[col] - user_vector[f"user_{col}"])

    X_pred = pd.DataFrame([row_dict])
    return X_pred[model_features]


def rank_users_for_track(track_id, df_full, user_profiles, best_model, feature_cols, model_features, top_n=None):
    """
    Gera um ranking de usuários mais propensos a ouvir uma música específica.

    Parameters
    ----------
    track_id : str
        ID da track a ser avaliada.
    df_full : pd.DataFrame
        DataFrame completo com as músicas (para buscar o track_id).
    user_profiles : pd.DataFrame
        Perfil médio de cada usuário (features prefixadas com 'user_').
    best_model : sklearn estimator
        Modelo treinado com predict_proba disponível.
    feature_cols : list
        Lista das colunas de features da música.
    model_features : list
        Ordem final das features usada no modelo.
    top_n : int, optional
        Número de usuários no topo do ranking a retornar.

    Returns
    -------
    pd.DataFrame
        DataFrame com colunas ['user', 'probabilidade'] ordenado desc.
    """
    # Buscar registros da música
    track_row = df_full[df_full["track_id"] == track_id]

    if track_row.empty:
        raise ValueError(f"Música não encontrada: {track_id}")

    # Características médias da música (caso apareça mais de uma vez)
    track_features = track_row[feature_cols].mean()

    ranking = []
    for user in user_profiles.index:
        user_vector = user_profiles.loc[user]
        X_pred = build_row_for_track_user(track_features, user_vector, feature_cols, model_features)
        prob = best_model.predict_proba(X_pred)[0][1]
        ranking.append((user, prob))

    ranking_df = (
        pd.DataFrame(ranking, columns=["user", "probabilidade"])
        .sort_values("probabilidade", ascending=False)
        .reset_index(drop=True)
    )

    if top_n is not None:
        ranking_df = ranking_df.head(top_n)

    return ranking_df
```

```python
# Exemplo de uso: ranking de usuários para um track_id específico
track_alvo = "5PaCrBjL6W6Gq60kJzuw9y"

ranking = rank_users_for_track(
    track_id=track_alvo,
    df_full=df,              
    user_profiles=user_profiles,
    best_model=best_model,
    feature_cols=feature_cols,
    model_features=model_features,
    top_n=5                
)

print(f"Ranking de usuários mais propensos a escutar a música {track_alvo}:")
ranking
```

## Avaliação do Modelo no Conjunto de Teste

Vamos avaliar o modelo usando:

- ROC AUC
- PR AUC (Average Precision)
- Accuracy
- Precision
- Recall
- F1-score
- Matriz de Confusão
- Classification Report completo

O limiar de classificação (threshold) é parametrizável (por padrão, 0.5).

```python
def evaluate_model(best_model, X_test, y_test, threshold=0.5):
    """
    Avalia o modelo em um conjunto de teste para um dado limiar de decisão.
    """
    y_pred_proba = best_model.predict_proba(X_test)[:, 1]
    y_pred = (y_pred_proba >= threshold).astype(int)

    roc_auc = roc_auc_score(y_test, y_pred_proba)
    pr_auc = average_precision_score(y_test, y_pred_proba)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    report = classification_report(y_test, y_pred)

    print(f"Limiar usado: {threshold}")
    print("ROC AUC:", roc_auc)
    print("PR AUC:", pr_auc)
    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1-score:", f1)

    print("\nMatriz de Confusão:")
    print(cm)

    print("\nRelatório completo:")
    print(report)

    return {
        "threshold": threshold,
        "roc_auc": roc_auc,
        "pr_auc": pr_auc,
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "confusion_matrix": cm,
        "report": report
    }
```

```python
# Rodar avaliação com threshold padrão 0.5
metrics = evaluate_model(best_model, X_test, y_test, threshold=0.5)
```

- heard = 1 → música escutada por pelo menos 60 segundos
- heard = 0 → música não escutada tempo suficiente

Também verificamos se existem as duas classes no dataset.


```python
df["heard"] = (df["ms_played"] >= 60000).astype(int)

print("Distribuição do label 'heard':")
print(df["heard"].value_counts(normalize=True))

if df["heard"].nunique() < 2:
    raise ValueError("Só há uma classe em 'heard'. Ajuste o threshold ou colete mais dados.")
```

## Seleção das Features Musicais

Selecionamos as variáveis acústicas que descrevem cada música.


```python
feature_cols = [
    "acousticness",
    "danceability",
    "energy",
    "instrumentalness",
    "liveness",
    "loudness",
    "speechiness",
    "valence"
]
```

## Split Treino/Teste

- 80% treino
- 20% teste
- Estratificação pela variável alvo


```python
df_train, df_test = train_test_split(
    df,
    test_size=0.2,
    random_state=42,
    stratify=df["heard"]
)
```

## Perfil do Usuário Ponderado por Tempo de Escuta

Criamos o perfil médio do usuário usando média ponderada:

- Peso = ms_played
- Usuários que escutam mais uma música influenciam mais o perfil
- Usamos apenas dados de treino para evitar vazamento de dados


```python
df_heard_train = df_train[df_train["heard"] == 1].copy()

def weighted_profile(x):
    return np.average(
        x[feature_cols],
        axis=0,
        weights=x["ms_played"]
    )

if df_heard_train.empty:
    raise ValueError("Nenhum registro com heard=1 no treino.")

user_profiles = (
    df_heard_train
    .groupby("user_name")
    .apply(weighted_profile)
)

user_profiles = pd.DataFrame(
    user_profiles.tolist(),
    index=user_profiles.index,
    columns=[f"user_{c}" for c in feature_cols]
)
```

## Engenharia de Features

- Juntamos o perfil do usuário ao treino e teste
- Criamos variáveis de diferença absoluta
- Construímos o conjunto final de features


```python
df_train_model = df_train.merge(
    user_profiles,
    left_on="user_name",
    right_index=True,
    how="inner"
)

df_test_model = df_test.merge(
    user_profiles,
    left_on="user_name",
    right_index=True,
    how="inner"
)

if df_test_model.empty:
    raise ValueError("Nenhum usuário do teste possui perfil calculado.")

for col in feature_cols:
    df_train_model[f"diff_{col}"] = abs(
        df_train_model[col] - df_train_model[f"user_{col}"]
    )
    df_test_model[f"diff_{col}"] = abs(
        df_test_model[col] - df_test_model[f"user_{col}"]
    )

model_features = (
    feature_cols +
    [f"user_{c}" for c in feature_cols] +
    [f"diff_{c}" for c in feature_cols]
)

X_train = df_train_model[model_features]
y_train = df_train_model["heard"]

X_test = df_test_model[model_features]
y_test = df_test_model["heard"]
```

## Treinamento do Random Forest

Configuração:

- 400 árvores
- min_samples_leaf = 5
- class_weight = balanced
- Paralelização ativada


```python
rf = RandomForestClassifier(
    n_estimators=400,
    min_samples_leaf=5,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)
```

## Ajuste do Threshold

Como estamos trabalhando com probabilidades, buscamos:

- Melhor threshold para F1-score
- Melhor recall mantendo precision >= 0.70


```python
def find_best_threshold(y_true, y_proba, metric="f1", min_precision=None):
    thresholds = np.linspace(0.1, 0.9, 81)
    best_thr = None
    best_score = -1
    best_metrics = None

    for thr in thresholds:
        y_pred = (y_proba >= thr).astype(int)
        prec = precision_score(y_true, y_pred, zero_division=0)
        rec = recall_score(y_true, y_pred, zero_division=0)
        f1 = f1_score(y_true, y_pred, zero_division=0)

        if min_precision is not None and prec < min_precision:
            continue

        score = {"f1": f1, "recall": rec, "precision": prec}[metric]

        if score > best_score:
            best_score = score
            best_thr = thr
            best_metrics = (prec, rec, f1)

    return best_thr, best_score, best_metrics
```

```python
probs = rf.predict_proba(X_test)[:, 1]

best_thr_f1, best_f1, (prec_f1, rec_f1, f1_f1) = find_best_threshold(
    y_test, probs, metric="f1"
)

print("Melhor threshold (F1):", best_thr_f1)
print(f"Precision: {prec_f1:.4f} | Recall: {rec_f1:.4f} | F1: {f1_f1:.4f}")
```

## Avaliação do Modelo

Calculamos:

- ROC AUC
- PR AUC
- Accuracy
- Precision
- Recall
- F1-score
- Matriz de Confusão
- Classification Report


```python
auc = roc_auc_score(y_test, probs)
pr_auc = average_precision_score(y_test, probs)

print("ROC AUC:", auc)
print("PR AUC:", pr_auc)

threshold = best_thr_f1
y_pred = (probs >= threshold).astype(int)

print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, zero_division=0))
print("Recall:", recall_score(y_test, y_pred, zero_division=0))
print("F1-score:", f1_score(y_test, y_pred, zero_division=0))

print("\nMatriz de Confusão:")
print(confusion_matrix(y_test, y_pred))

print("\nRelatório completo:")
print(classification_report(y_test, y_pred, zero_division=0))
```

## Ranking Personalizado de Usuários

Dado um track_id específico:

- Calculamos suas características médias
- Comparamos com cada perfil de usuário
- Geramos a probabilidade de cada usuário escutar
- Ordenamos do mais provável para o menos provável


```python
track_alvo = "5PaCrBjL6W6Gq60kJzuw9y"

track_row = df[df["track_id"] == track_alvo]
if track_row.empty:
    raise ValueError("Música não encontrada.")

track_features = track_row[feature_cols].mean()

ranking = []

for user in user_profiles.index:
    user_vector = user_profiles.loc[user]

    row_dict = {}
    for col in feature_cols:
        row_dict[col] = track_features[col]
        row_dict[f"user_{col}"] = user_vector[f"user_{col}"]
        row_dict[f"diff_{col}"] = abs(
            track_features[col] - user_vector[f"user_{col}"]
        )

    X_pred = pd.DataFrame([row_dict])
    X_pred = X_pred[model_features]

    prob_user = rf.predict_proba(X_pred)[0][1]
    ranking.append((user, prob_user))

ranking = (
    pd.DataFrame(ranking, columns=["user", "probabilidade"])
    .sort_values(by="probabilidade", ascending=False)
)

print("\nRanking de usuários para a track alvo:")
print(ranking)
```

## Feature Importance - Random Forest

```python
from sklearn.inspection import permutation_importance
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

result = permutation_importance(
    rf,
    X_test,
    y_test,
    n_repeats=10,
    random_state=42,
    n_jobs=-1
)

importances_mean = result.importances_mean
importances_std = result.importances_std

fi_df = (
    pd.DataFrame({
        "feature": model_features,
        "importance_mean": importances_mean,
        "importance_std": importances_std
    })
    .sort_values("importance_mean", ascending=False)
    .reset_index(drop=True)
)
fi_df_full = fi_df.copy()

fi_df_full["importance_model"] = -fi_df_full["importance_mean"]
fi_df_full["importance_model_std"] = fi_df_full["importance_std"]

fi_df_full["impact"] = fi_df_full["importance_model"].abs()

fi_df_full["effect"] = np.where(
    fi_df_full["importance_model"] > 0,
    "ajuda",       # positiva
    "atrapalha"    # negativa
)
fi_df_full = fi_df_full.sort_values("impact", ascending=False).reset_index(drop=True)

print(fi_df_full.head(20))
top_n = 25  

fi_top = fi_df_full.head(top_n).sort_values("importance_model")

colors = fi_top["importance_model"].apply(
    lambda x: "tab:green" if x > 0 else "tab:red"
)

plt.figure(figsize=(8, 10))
plt.barh(
    fi_top["feature"],
    fi_top["importance_model"],
    xerr=fi_top["importance_model_std"],
    color=colors,
    ecolor="black",
    align="center"
)
plt.axvline(0, color="black", linewidth=1)
plt.xlabel("Efeito no modelo (positivo = ajuda, negativo = atrapalha)")
plt.title("Importância das Variáveis - Random Forest")
plt.grid(axis="x", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
top_ajuda = fi_df_full[fi_df_full["importance_model"] > 0].head(15)
top_atrapalha = fi_df_full[fi_df_full["importance_model"] < 0].head(15)

print("\nFeatures que ajudam o modelo (positivas):")
print(top_ajuda[["feature", "importance_model", "impact"]])

print("\nFeatures que bagunçam o modelo (negativas):")
print(top_atrapalha[["feature", "importance_model", "impact"]])
```
