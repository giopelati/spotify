## Etapa 0: **Enriquecimento de Dados - API ReccoBeats**
- Integrar dados de características musicais da API ReccoBeats.
- Processar dados em paralelo utilizando Apache Spark.
- Consolidar informações de áudio com o dataset limitado do notebook anterior.

```python
import requests
from pyspark.sql.functions import udf 
from pyspark.sql.types import StringType 
from unidecode import unidecode
from pyspark.sql.functions import col
from requests.adapters import HTTPAdapter 
from urllib3.util.retry import Retry
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DateType
from time import sleep
```

### Importações e Configurações Iniciais
Importar as bibliotecas necessárias para manipulação de dados no Spark, requisições HTTP e processamento de strings.

```python
spark = SparkSession.builder \
    .master('local[*]') \
    .appName('api_calling') \
    .getOrCreate()
```

### Inicialização da Sessão Spark
Configurar a sessão Spark em modo local para processamento paralelo dos dados.

```python
schema_df = StructType([
    StructField('id',IntegerType(), nullable=False),
    StructField('ts', DateType(),nullable=True),
    StructField('ms_played',IntegerType(),nullable=True),
    StructField('track_name',StringType(),nullable=True),
    StructField('artist_name',StringType(),nullable=True),
    StructField('album_name',StringType(),nullable=True),
    StructField('user_name',StringType(),nullable=True),
    StructField('track_id',StringType(),nullable=True),
])
```

### Carregamento do Dataset Limitado
Definir o schema dos dados e carregá-los do arquivo CSV gerado no notebook anterior.

```python
folder = "./datasets"
temp_folder = "./datasets/temp_csv"
destiny = "./datasets/return.csv"


df = spark.read \
    .format("csv") \
    .schema(schema_df) \
    .option("header","false") \
    .option("encoding","utf-8") \
    .load(f"{folder}/limited_dataset.csv")

df.show(5)
```

```python
df = df.dropna()

df_grouped = (
    df.groupBy("track_id")
    .agg(
        F.first("track_name").alias("track_name")
    )
)
df_grouped.show(5)

df_grouped = df_grouped.repartition(36)
```

### Preparação dos Dados para Enriquecimento
- Remover linhas com valores ausentes.
- Agrupar por track_id e extrair informações únicas.
- Repartição para processamento paralelo eficiente.

```python
url_get_id = "https://api.reccobeats.com/v1/track?ids=:id"
url_get_info = "https://api.reccobeats.com/v1/track/:id/audio-features"
payload = {}

headers = {
    'Accept': 'application/json'
}
```

### Configuração das Requisições à API ReccoBeats
Definir URLs e headers necessários para comunicação com a API de características musicais.

```python
track_ids = df.select("track_id").collect()
session = requests.Session()
results = []
retries = Retry(total=5, backoff_factor=1, status_forcelist=[500, 502, 503, 504])
session.mount("https://", HTTPAdapter(max_retries=retries))


def process_partition(partition):
    for row in partition:
        track_id = row["track_id"]

        # 1. Buscar ID do Racoon
        url_get_track_id = url_get_id.replace(":id", track_id)
        resp = session.get(url_get_track_id, headers=headers, data=payload,verify=True).json()
        print(resp)
        if resp.get("content"):
            racoon_track_id = resp["content"][0]["id"]

            # 2. Buscar parâmetros musicais
            url_get_track_info = url_get_info.replace(":id", racoon_track_id)
            info_resp = session.get(url_get_track_info, headers=headers, data=payload, verify=True)
            print(info_resp.content)

            if info_resp.status_code == 200:
                info = info_resp.json()
                results.append((
                    track_id,
                    info.get("acousticness"),
                    info.get("danceability"),
                    info.get("energy"),
                    info.get("instrumentalness"),
                    info.get("liveness"),
                    info.get("loudness"),
                    info.get("speechiness"),
                    info.get("valence")
                ))
            else: 
                results.append((track_id, None, None, None, None, None, None, None, None)) 
        else: 
            results.append((track_id, None, None, None, None, None, None, None, None))
        sleep(1)

    return results
```

### Processamento Distribuído - Consulta à API
- Implementar função de processamento distribuído para cada partição.
- Buscar ID do Raccoon a partir do Spotify Track ID.
- Extrair características musicais: acousticness, danceability, energy, etc.
- Implementar retry automático para falhas de conexão.

```python
df_out = df_grouped.rdd.mapPartitions(process_partition).toDF([
    "track_id","acousticness","danceability","energy",
    "instrumentalness","liveness","loudness","speechiness","valence"
])

df_out.show(5)
```

### Criação do DataFrame com Características Musicais
Transformar resultados do processamento distribuído em um DataFrame estruturado.

```python
df_completed = df.join(df_out,on="track_id",how="inner")

df_completed.show()
```

### Consolidação de Dados - Join com Características
Unir o dataset original com as características musicais obtidas da API.

```python
df_completed = df_completed.drop("id")
df_completed.show()
```

### Limpeza de Colunas Desnecessárias
Remover coluna de índice que não é necessária para análises futuras.

```python
# Criar UDF para remover acentos 
unidecode_udf = udf(lambda s: unidecode(s) if s is not None else None, StringType()) 
columns_to_clean = ['track_name', 'artist_name', 'album_name'] 
df_limpo = df_completed.select([ unidecode_udf(col(c)).alias(c) if c in columns_to_clean else col(c) for c in df_completed.columns ])
df_limpo.show(20)
```

### Normalização de Texto - Remoção de Acentos
Aplicar UDF para remover acentos e caracteres especiais das colunas de texto (track_name, artist_name, album_name).

```python
df_limpo.coalesce(1).write.mode("overwrite").csv("return.csv", header=True)
```
