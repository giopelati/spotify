## Importação de Bibliotecas

Nesta etapa importamos bibliotecas necessárias para:

- **Processamento assíncrono** (`asyncio`, `aiohttp`)
- **Manipulação de dados** (`pandas`)
- **Controle de tempo e retries**
- **Manipulação de arquivos**


```python
import aiohttp
import asyncio
import pandas as pd
import time
import random
import os
from aiohttp import ClientTimeout, TCPConnector
```


## Configurações do Pipeline

Definimos parâmetros importantes para execução:

- **folder** → pasta onde os datasets estão armazenados  
- **url_get_id** → endpoint da API que retorna o ID interno da música  
- **url_get_info** → endpoint que retorna as *audio features*  
- **CONCURRENCY** → número máximo de requisições simultâneas


```python
folder = "../datasets"
url_get_id = "https://api.reccobeats.com/v1/track?ids=:id"
url_get_info = "https://api.reccobeats.com/v1/track/:id/audio-features"

CONCURRENCY = 100
```


## Função de Requisição Robusta

A função `robust_get` realiza chamadas HTTP de forma **resiliente**.

Ela inclui:

- **Retry automático**
- **Backoff exponencial**
- Tratamento de erros de rede
- Tratamento de **rate limit (429)**

Isso evita que o pipeline falhe durante chamadas massivas à API.


```python
async def robust_get(session, url):
    base = 0.5
    attempt = 0
    while True:  # loop infinito até conseguir resposta válida
        attempt += 1
        try:
            async with session.get(url) as resp:
                status = resp.status
                if 200 <= status < 300:
                    try:
                        return await resp.json(content_type=None)
                    except Exception:
                        return None
                if status in (429, 503):
                    ra = resp.headers.get("Retry-After")
                    if ra:
                        try:
                            delay = float(ra)
                        except ValueError:
                            delay = base * (2 ** attempt) + random.random() * 0.25
                    else:
                        delay = base * (2 ** attempt) + random.random() * 0.25
                    await asyncio.sleep(delay)
                    continue
                if 500 <= status < 600:
                    delay = base * (2 ** attempt) + random.random() * 0.25
                    await asyncio.sleep(delay)
                    continue
                return None
        except (aiohttp.ClientConnectionError, aiohttp.ClientOSError,
                aiohttp.ServerTimeoutError, asyncio.TimeoutError) as e:
            delay = base * (2 ** attempt) + random.random() * 0.25
            await asyncio.sleep(delay)
            continue
        except Exception as e:
            delay = base * (2 ** attempt) + random.random() * 0.25
            await asyncio.sleep(delay)
            continue
```


## Extração do ID da Música

A API pode retornar estruturas JSON diferentes.

A função `extract_id` garante que conseguimos recuperar o **ID interno da música** independentemente da estrutura retornada.


```python
def extract_id(resp_json):
    if not isinstance(resp_json, dict):
        return None
    content = resp_json.get("content")
    if isinstance(content, list) and content:
        first = content[0]
        if isinstance(first, dict):
            return first.get("id")
    if isinstance(content, dict):
        return content.get("id")
    return resp_json.get("id")
```


## Extração das Audio Features

A função `extract_features` coleta as principais características musicais retornadas pela API:

- acousticness
- danceability
- energy
- instrumentalness
- liveness
- loudness
- speechiness
- valence


```python
def extract_features(data):
    keys = ["acousticness","danceability","energy",
            "instrumentalness","liveness","loudness",
            "speechiness","valence"]
    def get_nested(d, k, default=None):
        if not isinstance(d, dict):
            return default
        if k in d:
            return d.get(k, default)
        c = d.get("content")
        if isinstance(c, dict):
            return c.get(k, default)
        return default
    return [get_nested(data, k, None) for k in keys]
```


## Processamento de uma Música

A função `process_track` executa o fluxo completo para uma música:

1. Buscar o **ID interno da música**
2. Consultar as **audio features**
3. Retornar todas as informações em formato estruturado


```python
async def process_track(session, track_id, idx):
    url_id = url_get_id.replace(":id", str(track_id))
    data_id = await robust_get(session, url_id)
    raccoon_id = extract_id(data_id)
    if not raccoon_id:
        return (track_id,) + (None,) * 8
    url_feat = url_get_info.replace(":id", str(raccoon_id))
    data_feat = await robust_get(session, url_feat)
    feats = extract_features(data_feat) if data_feat else [None] * 8
    return (track_id, *feats)
```


## Pipeline Principal

Esta etapa executa o fluxo completo:

1. Carregar o dataset
2. Agrupar músicas por `track_id`
3. Verificar cache existente
4. Identificar músicas ainda não processadas
5. Executar requisições assíncronas em paralelo
6. Salvar os resultados


```python
async def main():
    try:
        df = pd.read_csv(f"{folder}/limited_ts_dataset.csv", delimiter=',').dropna()
        df_grouped = df.groupby('track_id', as_index=False).agg(
            track_name=('track_name', lambda s: ','.join(sorted(set(s.dropna()))))
        )

        track_ids = df_grouped["track_id"].astype(str).tolist()

        cache_file = f"{folder}/music_info.csv"
        expected_cols = [
            'track_id','acousticness','danceability','energy',
            'instrumentalness','liveness','loudness','speechiness','valence'
        ]

        if os.path.exists(cache_file):
            df_cache = pd.read_csv(cache_file)
            df_cache = df_cache.reindex(columns=expected_cols, fill_value=None)
            done_ids = {
                row["track_id"]: tuple(row[col] for col in expected_cols)
                for _, row in df_cache.iterrows()
                if not pd.isnull(row["acousticness"])
            }
        else:
            df_cache = pd.DataFrame(columns=expected_cols)
            done_ids = {}

        track_ids_new = [tid for tid in track_ids if tid not in done_ids]


        timeout = ClientTimeout(total=None, connect=None, sock_read=None)
        connector = TCPConnector(limit=CONCURRENCY, limit_per_host=CONCURRENCY, ttl_dns_cache=300,ssl=False)
        headers = {"Accept":"application/json","User-Agent":"reccobeats-client/1.0"}

        async with aiohttp.ClientSession(timeout=timeout, connector=connector, headers=headers) as session:
            start = time.time()
            results = []
            sem = asyncio.Semaphore(CONCURRENCY)
            async def worker(tid, idx):
                async with sem:
                    return await process_track(session, tid, idx)
            tasks = [worker(tid, i) for i, tid in enumerate(track_ids_new)]
            for coro in asyncio.as_completed(tasks):
                res = await coro
                results.append(res)
    except KeyboardInterrupt:
        df_out = pd.DataFrame(results, columns=expected_cols)
        df_final = pd.concat([df_cache, df_out], ignore_index=True)

        df_final = df_final.dropna()
        df_final.to_csv(cache_file, index=False)


        
    finally:
        df_out = pd.DataFrame(results, columns=expected_cols)
        df_final = pd.concat([df_cache, df_out], ignore_index=True)

        df_final = df_final.dropna()
        df_final.to_csv(cache_file, index=False)
```


## Junção dos dataframes

Realiza um join com as tabelas de músicas, contendo as informações e os detalhes da música 


```python
def join_tracks():
    df_limited_dataset_ts = pd.read_csv(f"{folder}/limited_ts_dataset.csv", delimiter=',').dropna()
    df_music_info = pd.read_csv(f"{folder}/music_info.csv", delimiter=',').dropna()

    # Corrigir tipo da coluna
    df_limited_dataset_ts['track_id'] = df_limited_dataset_ts['track_id'].astype(str)
    df_music_info['track_id'] = df_music_info['track_id'].astype(str)

    display(df_music_info)
    display(df_limited_dataset_ts)

    joined_df = df_limited_dataset_ts.merge(df_music_info, on='track_id', how='inner')

    joined_df.to_csv(f"{folder}/return_corrected.csv", index=False)
```


## Execução do Pipeline

Por fim executamos o processamento assíncrono.


```python
await main()
```
